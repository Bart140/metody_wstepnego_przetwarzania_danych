## Zadanie 2

```python
index = ["base"]
for i in range(2,11):
    index.append(f"pow_{i}")
```

```python
numbers = np.vander(np.arange(1,11),11,increasing = True)
numbers = numbers[:,1:]
```

```python
print(numbers)
```

```python
df = pd.DataFrame(numbers, index = index)
print(len(index),len(numbers))
```

```python
print(df)
```

## Zadanie 3

```python
print(np.shares_memory(numbers,df))
```

```python
df_shared = pd.DataFrame(numbers, index = index, copy = False)
```

```python
print(np.shares_memory(numbers,df_shared))
```

## Zadanie 4

```python
print(df.dtypes.head)
memory_usage = df_shared.memory_usage(deep=True).sum()
```

```python
print(memory_usage)
```

```python
df_downcast = df.apply(pd.to_numeric,downcast = "integer")
```

```python
print(memory_usage_downcasted)
```
